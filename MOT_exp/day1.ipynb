```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scienceplots 
plt.style.use(['science', 'ieee'])
```

```python
# Import the data from csv 

cooling_spec_path = r'Data\Day1\Cooling_spec.csv'



# Convert the data to numpy arrays
data = pd.read_csv(cooling_spec_path, header=0).to_numpy()
print(data)
time = data[:, 0]
ch1 = data[:, 1]
ch2 = data[:, 2]
ch3 = data[:, 3]

# Find peaks in the data with filtering for obvious peaks
from scipy.signal import find_peaks
peaks_ch2, _ = find_peaks(ch2, height=0.1, distance=50)
peaks_ch3, _ = find_peaks(ch3, height=0.1, distance=50)

fig, ax = plt.subplots()
ax.plot(time, ch2, label='Channel 2', color='orange')
ax.plot(time, ch3, label='Channel 3', color='green')

# Annotate peaks for Channel 2
for peak in peaks_ch2:
    ax.annotate(f'{ch2[peak]:.2f}', (time[peak], ch2[peak]), textcoords="offset points", xytext=(0,10), ha='center', color='orange')

# Annotate peaks for Channel 3
for peak in peaks_ch3:
    ax.annotate(f'{ch3[peak]:.2f}', (time[peak], ch3[peak]), textcoords="offset points", xytext=(0,10), ha='center', color='green')

ax.set_xlabel('Time (s)')
ax.set_ylabel("Oscilloscope Voltage (V)")
ax.grid()
ax.set_xlim(-0.0005,0.002)
ax.legend()
ax.set_title("Cooling Laser Spectrum")
```
